```python
#Task 1: Registration & Login Using File Handling

#Email Check
import re
def Email_Check():
  c = 8

  while c != 0:
    name = input("Enter email: ")  #Eg: abc123@gmail.com
    regex = r'\b[a-zA-Z]+[a-zA-Z0-9]+[@]+[a-zA-Z]+[.]+[a-zA-z]{2,}'

    if(re.fullmatch(regex, name)):
        print("\nValid Email\n")
        return name
    else:
        print("\nInvalid Email !! Re-enter Email\n")
        c=1

#Password Check:
import re
def Password_Check():
  c = 8

  while c != 0:
      passw = input("Enter Password: ")  #Eg: Abc@123
      
      regex = r'\b(?=.*[a-z])(?=.*[A-Z])(?=.*[0-9])(?=.*[!@#$%^&*-.])'

      if(re.findall(regex, passw)):
          print("\nValid Password \nUser Registered Successfully")
          return passw
          c=0
      else:
          print("\nInvalid Password !! Re-enter Password\n")
          c=1

#Login
def Login():
  #name = input("\nEnter Username: ")
  name = Email_Check()
  passw = input("Enter Password: ")
    
  f = open("User_Data.txt",'r')
  i = f.read()
  i = i.split()
  if name in i:
      index = i.index(name) + 1
      usr_password = i[index]
      if usr_password == passw:
          print("\nUser Found !! \nWelcome,", name)
      else:
          print("\nWrong Password !!\n")

          n = int(input("Enter 1.Forgot Password  0.Exit :"))
          if n == 1:
            index = i.index(name) + 1
            print("Password  ==> ",i[index])
          elif n == 0:
            print("\nExit ")
          else:
            print("\nInvalid Entry !!")

  else:
      print("User not found. Please Register.") 

#Register
def Register():
  #name = input("\nEnter Username: ")
  name = Email_Check()
  #passw = input("Enter Password: ")
  passw = Password_Check()
  f = open("User_Data.txt",'r')
  i = f.read()
  if name in i:
    print("Username not available. Please try using new username")
    f.close()
  else:
    f.close()
    f = open("User_Data.txt",'w')
    i = i + " " +name + " " + passw
    f.write(i)

#Main() Program
c=8
while c!=0:
  c = int(input("\nEnter  1.Login   2.Registration  0.Exit : "))
  if c == 1:
    Login()
  elif c == 2:
    Register()
  elif c == 0:
    print("\nExit \nThank You !!")
  else:
    print("\nInvalid Entry !!")
```
